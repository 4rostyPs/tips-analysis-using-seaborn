import modules

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

load tips dataset

```python
tips = sns.load_dataset("tips")
```

use head to show first 5 rows

```python
tips.head()
```

Show mean of dataset collumns which contain numerical data

```python
tips.mean()
```

creating series

```python
tip = tips["tip"]
```

mean of tip series

```python
tip.mean()
```

median of tip series

```python
tip.median()
```

histplot for tip collumn

```python
sns.histplot(data = tips,x = "tip")
```

Average tip is between 2 and 4 dollars

histplot for total bill column

```python
sns.histplot(data = tips,x="total_bill")
```

Average bill is between 10 and 20 dollars

```python
tips["total_bill"].mean()
```

```python
tips["total_bill"].mode()
```

kdeplot for size collumn

```python
sns.kdeplot(data=tips,x="size")
```

Most common tip is about 2 dollars

kdeplot for total bill column

```python
sns.kdeplot(data=tips,x="total_bill")
```

Most common bill is about 15 dollars

kde plot for tip column

```python
sns.kdeplot(data=tips,x="tip")
```

Most common tip is about 2 ollars and most tips lie between 1 and 4 dollars 

comparision of tips given by smokers vs non smokers. hue is used for colour encoding to differentiate between smokers and non smokers.

```python
sns.kdeplot(data=tips,x="tip",hue="smoker")
```

Smokers and non-smokers generally tip the same

```python
sns.kdeplot(data=tips,x="tip",hue="sex")
```

Males and females generally tip the same

```python
sns.kdeplot(data=tips,x="total_bill",hue="smoker")
```

Smokers and non-smokers generally have simmilar total bills

using box plot to compare the tips that males give vs. the tips that females give

```python
sns.boxplot(data=tips,x="tip",y="sex")
```

Males geerally leave higher tips and have more exceptional cases of high tips

```python
sns.boxplot(data=tips,x="tip",y="smoker")
```

Non-smokers tip slightly less than smokers (based on median) but have more exceptional cases of high tips

scatterplot comparing total bill and tip. used to find correlation if any.

```python
sns.scatterplot(data=tips,x="total_bill",y="tip")
```

scatterplot using hue for colour encoding to compare males and females tips.

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex")
```
